```python
import pandas as pd
train = pd.read_csv('train.csv')
features = pd.read_csv('features.csv')
stores = pd.read_csv('stores.csv')
```

```python
# Merge datasets
df = train.merge(features, on=['Store', 'Date', 'IsHoliday'], how='left')
df = df.merge(stores, on='Store', how='left')
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
### Aggregate Sales and Prepare Features

# Aggregate by date
sales_ts = df.groupby('Date')['Weekly_Sales'].sum().reset_index()
sales_ts.columns = ['ds', 'y']  # Rename for Prophet
```

```python
###check weekly sales
data=df.groupby('Date')['Weekly_Sales'].sum()
data.plot()
plt.xlabel(None)
plt.show()
```

```python
### Add Holiday and Promotion Effects

# Extract holiday dates
holidays = df[df['IsHoliday'] == True]['Date'].unique()

from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
cal = calendar()
holidays_df = cal.holidays(start=sales_ts['ds'].min(), end=sales_ts['ds'].max())

# Format holidays for Prophet
holiday_df = pd.DataFrame({
    'holiday': 'us_holiday',
    'ds': pd.to_datetime(holidays),
    'lower_window': 0,
    'upper_window': 1
})
```

```python
### Forecast with Prophet

from prophet import Prophet

model = Prophet(holidays=holiday_df, yearly_seasonality=True, weekly_seasonality=True)
model.fit(sales_ts)

# Future dataframe
future = model.make_future_dataframe(periods=12, freq='W')
forecast = model.predict(future)
```

```python
### Evaluate the Model

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Perform an inner join to ensure both 'y' and 'yhat' are aligned and have the same length
predicted = forecast[['ds', 'yhat']].merge(sales_ts[['ds', 'y']], on='ds', how='inner')

# Now calculate the error metrics
mae = mean_absolute_error(predicted['y'], predicted['yhat'])
rmse = np.sqrt(mean_squared_error(predicted['y'], predicted['yhat']))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

```python
 ###Visualize

model.plot(forecast);

# Optional: Component plots
model.plot_components(forecast);
```

```python
import seaborn as sns

### Create pivot table of average weekly sales by month and year
heatmap_data = sales_ts.copy()
heatmap_data['Year'] = heatmap_data['ds'].dt.year
heatmap_data['Month'] = heatmap_data['ds'].dt.month
heatmap_data['Week'] = heatmap_data['ds'].dt.isocalendar().week

pivot_table = heatmap_data.pivot_table(index='Week', columns='Month', values='y', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5)
plt.title('Average Weekly Sales Heatmap by Month')
plt.xlabel('Month')
plt.ylabel('Week')
plt.show()
```
